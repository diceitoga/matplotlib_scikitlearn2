```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
points = np.arange(-5, 5, 0.01)
print('Points: ')
print(points[:10])   #This only shows the first 10 points just as an example
```

```python
dx,dy = np.meshgrid(points, points)
```

```python
dx
```

```python
dy
```

```python
z = (np.sin(dx)+ np.sin(dy))

#plt.plot(z)
plt.imshow(z)
plt.colorbar()
plt.title("Addition of sin dx and sin dy")
```

# np where

```python
a = np.array([1,2,3,4])
b = np.array([100,200,300,400])
condition = np.array([True,True,False,False])
```

```python
answer = [(A_val if condition else B_val) for A_val, B_val, cond in zip(a,b,condition)]
```

```python
answer2 = np.where(condition,a,b)       #np.where(condition(True/False), A(if True),B(if False))
answer2
```
