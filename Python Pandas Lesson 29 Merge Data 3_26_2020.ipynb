```python
# Merge on columns

import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
dframe1 = DataFrame({'key': ['x','z','y','z','x','x'],'data_set_1':np.arange(6)})
```

```python
dframe1 
```

```python
dframe2 = DataFrame({'key':['q','y','z'],
                     'data_set_2':[1,2,3]})
```

```python
dframe2
```

# pd.merge(df1, df2)

```python
pd.merge(dframe1,dframe2)     # z and y are the only values available on both DFs 
```

```python
pd.merge(dframe1,dframe2,on='key')
```

```python
pd.merge(dframe1,dframe2,on='key',how ='left')
```

```python
pd.merge(dframe1,dframe2,on='key',how ='right')
```

```python
pd.merge(dframe1, dframe2, on='key',how ='outer')
```

```python
df_left = DataFrame({'key1': ['SF','SF','LF'],
                     'key2': ['One','Two','One'],
                     'left_data':[10,20,30]
                    })
```

```python
df_right = DataFrame({'key1':['SF','SF','LA','LA'],
                      'key2':['One','One','One','Two'],
                      'right_data':[40,50,60,70]
                     })
```

```python
print(df_left)
print(df_right)
```

```python
pd.merge(df_left,df_right,on=['key1','key2'], how ='outer')   #outer means union or OR
```

```python
pd.merge(df_left,df_right,on='key1')  
```
